# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

## Side-by-Side Graph

### Advantages
* You can clearly see each time series separately without overlap. This makes it easier to see the highs and lows for the week.
* It looks cleaner since the lines aren't overlapping.
* The museum staff can focus on one week at a time.

### Disadvantages
* It's hard to compare one day like Monday for week 1 against Monday for week 2. Your eyes have to jump back and forth.
* Requires double the space which could be an issue if you're trying to present the data.
* THe viewer needs to mentally combine the images in order to understand the weeks against each other.

## Overlapping Graph

### Advantages
* You can see immediately which days had higher and lower attendance between the different weeks.
* It's easy to see if overall attendance is increasing or decreasing.
* It's on a single plot which makes it better for presentations.

### Disadvantages
* If you have more than 2-3 weeks there would be too many lines and it would be hard to distinguish between the weeks.
* You need to clearly label the weeks.
* It's harder to analyze one week individually.

## Summary

Overall for this data I would pick the overlapping graph since it is easier to spot trends and we only have two weeks. The viewer can easily compare the two week against each other and spot trends.

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014])

product_a = [395, 415, 425, 390, 305, 270, 255]
product_b = [360, 395, 410, 380, 290, 255, 245]
product_c_years = [2010, 2011, 2012, 2013, 2014]
product_c = [100, 160, 190, 235, 180]
product_d_years = [2011, 2012, 2013, 2014]
product_d = [150, 245, 215, 215]
product_e_years = [2013, 2014]
product_e = [100, 210]

plt.figure(figsize=(10, 6))

plt.plot(years, product_a, 'b-', linewidth=3, label='Product A')
plt.plot(years, product_b, color='darkred', linewidth=3, label='Product B')
plt.plot(product_c_years, product_c, color='limegreen', linewidth=3, label='Product C')
plt.plot(product_d_years, product_d, color='blueviolet', linewidth=3, label='Product D')
plt.plot(product_e_years, product_e, color='cyan', linewidth=3, label='Product E')

plt.xlim(2007.5, 2014.5)
plt.ylim(0, 500)

plt.xlabel('')
plt.ylabel('')
plt.title('Average Retail Product Price per Year', fontsize=20, fontweight='bold', pad=20)

yticks = range(0, 600, 100)
plt.yticks(yticks, ['${}'.format(y) for y in yticks], fontsize=14)
plt.xticks(years, fontsize=14)

plt.grid(True, alpha=0.3)

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=5, fontsize=16, frameon=False)

plt.tight_layout()

plt.show()
```
