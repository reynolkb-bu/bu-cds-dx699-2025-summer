Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

Histograms Pros:
* You can easily see what drugs fall into which price range
* You can see the cheaper ones are more prevelant on the left side

Histogram Cons:
* You're not able to see specific statistics like the min, median or max.
* It's harder to see outlier data.

Boxplot Pros
* You're clearly able to see minimum, Q1, median, Q3, and maximum.
* You can see the outliers clearly which are the dots outside of the whiskers.

Boxplot Cons
* You're not able to see if there are a lot of values within a certain bin range.
* It's harder to glance at the data and see what is going on.

```python
df = pd.read_csv('Diabetic_Nephropathy_v1.csv')
print(f"Dataset shape: {df.shape}")
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace('（', '(').str.replace('）', ')')
plt.style.use('seaborn-v0_8-darkgrid')
categorical_cols = ['Sex', 'Diabetic retinopathy (DR)', 'Diabetic nephropathy (DN)', 'Smoking', 'Drinking', 'Insulin', 'Metformin', 'Lipid lowering drugs']
numerical_cols = ['Age', 'Diabetes duration (y)', 'Height(cm)', 'Weight(kg)', 'BMI (kg/m2)', 'SBP (mmHg)', 'DBP (mmHg)', 'HbA1c (%)', 'FBG (mmol/L)', 'TG(mmoll)', 'C-peptide (ng/ml)', 'TC(mmoll)', 'HDLC(mmoll)', 'LDLC(mmoll)']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Categorical Variables Distribution', fontsize=16)
axes = axes.flatten()
for idx, col in enumerate(categorical_cols):
    value_counts = df[col].value_counts().sort_index()

    if col == 'Sex':
        labels = ['Female', 'Male'] if 'F' in value_counts.index else value_counts.index
    else:
        labels = ['No', 'Yes']
        
    axes[idx].bar(range(len(value_counts)), value_counts.values,
                  tick_label=labels, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'{col} Distribution')
    axes[idx].set_ylabel('Count')

    for i, v in enumerate(value_counts.values):
        percentage = (v / len(df)) * 100
        axes[idx].text(i, v + 5, f'{percentage:.1f}%', ha='center')

# hide last subplot
axes[-1].axis('off')
plt.tight_layout(rect=(0, 0, 1, 0.98))
plt.show()
```

```python
fig, axes = plt.subplots(5, 4, figsize=(16, 20))
fig.suptitle('Numerical Variables Distribution (Age to BMI) - Multiple Visualization Types', fontsize=16)
plot_types = ['Histogram', 'KDE Plot', 'Box Plot', 'Violin Plot']

for row, col in enumerate(numerical_cols[:5]):
    data = df[col].dropna()

    axes[row, 0].hist(data, bins=30, edgecolor='black', alpha=0.7, color='skyblue')
    axes[row, 0].set_title(f'{col} - Histogram')
    axes[row, 0].set_xlabel(col)
    axes[row, 0].set_ylabel('Frequency')

    data.plot.kde(ax=axes[row, 1], color='darkblue')
    axes[row, 1].set_title(f'{col} - KDE Plot')
    axes[row, 1].set_xlabel(col)

    axes[row, 2].boxplot(data, vert=True, patch_artist=True)
    axes[row, 2].set_title(f'{col} - Box Plot')
    axes[row, 2].set_ylabel(col)

    parts = axes[row, 3].violinplot(data, vert=True, showmeans=True, showmedians=True)
    axes[row, 3].set_title(f'{col} - Violin Plot')
    axes[row, 3].set_ylabel(col)

plt.tight_layout(rect=(0, 0, 1, 0.98))
plt.show()
```

```python
fig, axes = plt.subplots(5, 4, figsize=(16, 20))
fig.suptitle('Numerical Variables Distribution (Blood Pressure & Lab Values) - Multiple Visualization Types', fontsize=16)

for row, col in enumerate(numerical_cols[5:10]):
    data = df[col].dropna()

    axes[row, 0].hist(data, bins=30, edgecolor='black', alpha=0.7, color='lightcoral')
    axes[row, 0].set_title(f'{col} - Histogram')
    axes[row, 0].set_xlabel(col)
    axes[row, 0].set_ylabel('Frequency')

    data.plot.kde(ax=axes[row, 1], color='darkred')
    axes[row, 1].set_title(f'{col} - KDE Plot')
    axes[row, 1].set_xlabel(col)

    box = axes[row, 2].boxplot(data, vert=True, patch_artist=True)
    box['boxes'][0].set_facecolor('lightcoral')
    axes[row, 2].set_title(f'{col} - Box Plot')
    axes[row, 2].set_ylabel(col)

    parts = axes[row, 3].violinplot(data, vert=True, showmeans=True, showmedians=True)
    axes[row, 3].set_title(f'{col} - Violin Plot')
    axes[row, 3].set_ylabel(col)

plt.tight_layout(rect=(0, 0, 1, 0.98))
plt.show()
```

```python
fig, axes = plt.subplots(4, 4, figsize=(16, 16))
fig.suptitle('Numerical Variables Distribution (Lipid Profile) - Multiple Visualization Types', fontsize=16)

for row, col in enumerate(numerical_cols[10:14]):
    data = df[col].dropna()

    axes[row, 0].hist(data, bins=30, edgecolor='black', alpha=0.7, color='lightgreen')
    axes[row, 0].set_title(f'{col} - Histogram')
    axes[row, 0].set_xlabel(col)
    axes[row, 0].set_ylabel('Frequency')

    data.plot.kde(ax=axes[row, 1], color='darkgreen')
    axes[row, 1].set_title(f'{col} - KDE Plot')
    axes[row, 1].set_xlabel(col)

    box = axes[row, 2].boxplot(data, vert=True, patch_artist=True)
    box['boxes'][0].set_facecolor('lightgreen')
    axes[row, 2].set_title(f'{col} - Box Plot')
    axes[row, 2].set_ylabel(col)

    parts = axes[row, 3].violinplot(data, vert=True, showmeans=True, showmedians=True)
    axes[row, 3].set_title(f'{col} - Violin Plot')
    axes[row, 3].set_ylabel(col)

plt.tight_layout(rect=(0, 0, 1, 0.98))
plt.show()
```
