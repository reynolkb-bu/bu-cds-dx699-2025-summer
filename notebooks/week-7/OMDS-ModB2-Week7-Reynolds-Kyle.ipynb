Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

If I wanted to cause the viewer to examine the last three months of data first I would make the changes below.

1. I would make the first 9 months a different color like a gray or light blue. Essentially a color that doesn't stand out. Then I'd make the last 3 months a bright color like red, orange or a bright blue. Since people's eyes are drawn to high contrast areas, they would immediately focus on the last 3 months.
2. I would make the last 3 months thicker so they stand out more. This is because thicker lines will cause the viewer to focus on them more. I might even play around with a gradual thickness transition.
3. I would add symbols/markers like circles on the last 3 months. These would act as anchors and require people's immediate attention.

Initially I would just try color and see how it looks. If I add too many changes, it might be overwhelming for the human eye.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

processed = [160, 180, 240, 150, 180, 150, 130, 150, 130, 120, 130, 140]
received = [160, 180, 240, 150, 180, 160, 140, 210, 170, 150, 160, 190]

plt.figure(figsize=(10, 6))
plt.plot(months, received, color='grey', label='Received', marker='o')
plt.plot(months, processed, color='blue', label='Processed', marker='o')

for i in range(7, 12):
    plt.text(months[i], received[i] + 5, str(received[i]), ha='center', va='bottom', fontsize=9)
    plt.text(months[i], processed[i] + 5, str(processed[i]), ha='center', va='bottom', fontsize=9)

plt.ylim(0, 300)
plt.yticks(range(0, 301, 50))
plt.legend(bbox_to_anchor=(1.02, 0.55), loc='center left')
plt.grid(True, alpha=0.3)

plt.show()
```
